This notebook allows the reproduce **Figure S2** of the paper "Covering the Combinatorial Design Space of Multiplex CRISPR/Cas Experiments in Plants" by Kirsten Van Huffel, Michiel Stock, Thomas Jacobs, Tom Ruttink and Bernard De Baets.

#### Load packages

```julia
using Random 
using Plots  
using Distributions 
using LinearAlgebra
using Combinatorics
using BioCCP
```

#### Load BioCCP- and simulation-based framework

```julia
include("MultiplexCrisprDOE.jl");
```

#### Set seed for reproducibility

```julia
Random.seed!(1)
```

#### Study relation between expected combinatorial coverage of triples and given plant library size 

```julia
include("DefaultParameters_k=3.jl");
```

- Figure S2A: expected combinatorial coverage w.r.t. plant library size

```julia
plant_library_sizes = 0:1000:40000
expected_γₓ₃ = []

for N in plant_library_sizes
    push!(expected_γₓ₃, BioCCP_γₓ₃(x, 
                                         N,
                                         g, 
                                         r, 
                                         n_gRNA_total, 
                                         p_gRNA_freq, 
                                         p_gRNA_edit, ϵ_KO))
end
 

plot(plant_library_sizes, expected_γₓ₃,
label="E[γₓ₃]", title="Expected combinatorial coverage w.r.t. plant library size 
    ",
xlabel="N", ylabel="E[γₓ₃]", 
        xticks = (0:5000:50000, string.(0:5000:50000)),
 size=(850,450), color=:turquoise4, linewidth=2)
hline!([0.95], linestyle=:dash, color=:grey, label="E[γₓ₃] = 0.95", legend=:bottomright)
```

#### Study relation between probibility of full combinatorial coverage and given plant library size 

- Figure S2B: probability of full combinatorial coverage w.r.t. plant library size

```julia
plant_library_sizes = 0:1000:80000
Pₓ₃ = []

for N in plant_library_sizes
    push!(Pₓ₃, BioCCP_Pₓ₃(x, 
                            N,
                            g, 
                            r, 
                            n_gRNA_total, 
                            p_gRNA_freq, 
                            p_gRNA_edit, 
                            ϵ_KO))
    
end
 
plot(plant_library_sizes, Pₓ₃, label="Pₓ₃", 
    title="Probability of full combinatorial coverage with respect to plant library size",
xlabel="N
    ", ylabel="
    Pₓ₃", 
        xticks = (0:10000:80000, string.(0:10000:80000)),
 size=(850,400), color=:turquoise4, linewidth=2)
hline!([0.95], linestyle=:dash, color=:grey, label="Pₓ₃ = 0.95", legend=:bottomright)
```
