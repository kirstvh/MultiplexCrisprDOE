- Figure 3A: Distribution of the abundances of the gRNAs in the construct library

```julia
Random.seed!(1)
p_gRNA_freq = gRNA_frequency_distribution(m, sd, l, u, n_gRNA_total; normalize = false, visualize=false)

histogram(p_gRNA_freq,
         label="", xlabel="Number of reads per gRNA", 
         linecolor = "white", 
        normalize = :probability, 
         xtickfontsize=10,ytickfontsize=10, 
            xlim = (25, 150), 
    color=:mediumturquoise, size=(600,350),
    bins = 25,
            ylabel="relative frequency", title="Read distribution")
```

- Figure 3B: Distribution of genome editing efficiencies of the gRNAs
